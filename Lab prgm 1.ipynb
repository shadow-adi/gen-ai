```python
import requests

API_KEY = "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"
HEADERS = {"Authorization": f"Bearer {API_KEY}"}

MODELS = {
    "sentiment": "cardiffnlp/twitter-roberta-base-sentiment",
    "summarizer": "facebook/bart-large-cnn",
}


def query(model, payload):
    url = f"https://router.huggingface.co/hf-inference/models/{model}"
    try:
        r = requests.post(url, headers=HEADERS, json=payload)
        if r.status_code != 200:
            print(f"API error {r.status_code}: {r.text}")
            return None
        return r.json()
    except Exception as e:
        print(f"Request failed: {e}")
        return None


def summarize_prompt(text):
    out = query(MODELS["summarizer"], {"inputs": text})
    if out and isinstance(out, list):
        return out[0].get("summary_text", "")
    return "Failed to generate response."


def generate_story(theme):
    prompt = (
        f"Once upon a time, there was a magical land where {theme} was important. "
        f"Write a fun children's story about this."
    )
    return summarize_prompt(prompt)


def generate_interview_questions(role):
    prompt = (
        f"Generate at least 5 professional interview questions for the role of {role}. "
        f"Keep them clear and concise."
    )
    return summarize_prompt(prompt)


def generate_startup_idea(domain):
    prompt = (
        f"Suggest innovative, short and creative startup ideas in the field of {domain}."
    )
    return summarize_prompt(prompt)


def analyze_sentiment(text):
    out = query(MODELS["sentiment"], {"inputs": text})
    if out and isinstance(out, list):
        return out[0]
    return "Failed to analyze sentiment."



def main():
    options = {
        "1": ("Story Generator", lambda: generate_story(input("Enter a theme: "))),
        "2": ("Interview Question Generator", lambda: generate_interview_questions(input("Enter a job role: "))),
        "3": ("Sentiment Analyzer", lambda: analyze_sentiment(input("Enter text: "))),
        "4": ("Text Summarizer", lambda: summarize_prompt(input("Enter text: "))),
        "5": ("Startup Idea Generator", lambda: generate_startup_idea(input("Enter a domain/industry: "))),
    }

    while True:
        print("\nAI Multi-Tool Assistant")
        print("1. Story Generator")
        print("2. Interview Question Generator")
        print("3. Sentiment Analyzer")
        print("4. Text Summarizer")
        print("5. Startup Idea Generator")
        print("0. Exit")

        choice = input("Choose an option: ").strip()

        if choice == "0":
            print("Exiting...")
            break

        action = options.get(choice)
        if action:
            label, func = action
            print(f"\n{label} Result:\n")
            print(func())
        else:
            print("Invalid choice. Try again.")

if __name__ == "__main__":
    main()
```
