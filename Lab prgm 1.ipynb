```python
import requests

# 🔑 Replace with your actual Hugging Face API key
API_KEY = "hf_UOLQOMkJPWHEgOJVbRvYRWaO"
HEADERS = {"Authorization": f"Bearer {API_KEY}"}

# Models (only those available free on Inference API)
MODELS = {
    "sentiment": "cardiffnlp/twitter-roberta-base-sentiment",  # works
    "summarizer": "facebook/bart-large-cnn"                     # works
}

def query(model, payload):
    url = f"https://api-inference.huggingface.co/models/{model}"
    try:
        response = requests.post(url, headers=HEADERS, json=payload)
        if response.status_code != 200:
            print(f"⚠️ API Error {response.status_code}: {response.text}")
            return None
        return response.json()
    except Exception as e:
        print(f"❌ Request failed: {str(e)}")
        return None

# Story generator (hack using summarizer creatively)
def generate_story(theme):
    text = f"Once upon a time, there was a magical land where {theme} played a big role. Write a fun children's story about this."
    output = query(MODELS["summarizer"], {"inputs": text, "parameters": {"max_length": 120}})
    if output and isinstance(output, list):
        return output[0]["summary_text"]
    return "⚠️ Failed to generate story."

# Interview questions (hack)
def generate_interview_questions(role):
    text = f"Generate professional interview questions for the role of {role}. Include at least 5 questions."
    output = query(MODELS["summarizer"], {"inputs": text, "parameters": {"max_length": 120}})
    if output and isinstance(output, list):
        return output[0]["summary_text"]
    return "⚠️ Failed to generate questions."

# Startup ideas
def generate_startup_idea(domain):
    text = f"Suggest innovative startup ideas in the field of {domain}. Keep them short and creative."
    output = query(MODELS["summarizer"], {"inputs": text, "parameters": {"max_length": 120}})
    if output and isinstance(output, list):
        return output[0]["summary_text"]
    return "⚠️ Failed to generate ideas."

# Sentiment analysis
def analyze_sentiment(text):
    output = query(MODELS["sentiment"], {"inputs": text})
    if output:
        return output[0]
    return "⚠️ Failed to analyze sentiment."

# Summarizer
def summarize_text(text):
    output = query(MODELS["summarizer"], {"inputs": text, "parameters": {"max_length": 100}})
    if output and isinstance(output, list):
        return output[0]["summary_text"]
    return "⚠️ Failed to summarize."

# === MAIN PROGRAM ===
def main():
    while True:
        print("\n🤖 AI Multi-Tool Assistant ===")
        print("1. Story Generator")
        print("2. Interview Question Generator")
        print("3. Sentiment Analyzer")
        print("4. Text Summarizer")
        print("5. Startup Idea Generator")
        print("0. Exit")

        choice = input("Choose an option: ")

        if choice == "1":
            theme = input("Enter a theme for the story: ")
            print("\n📖 Story:\n", generate_story(theme))

        elif choice == "2":
            role = input("Enter a job role: ")
            print("\n❓ Interview Questions:\n", generate_interview_questions(role))

        elif choice == "3":
            text = input("Enter text to analyze: ")
            print("\n😊 Sentiment:\n", analyze_sentiment(text))

        elif choice == "4":
            text = input("Enter text to summarize: ")
            print("\n📝 Summary:\n", summarize_text(text))

        elif choice == "5":
            domain = input("Enter a domain/industry: ")
            print("\n🚀 Startup Ideas:\n", generate_startup_idea(domain))

        elif choice == "0":
            print("👋 Exiting... Goodbye!")
            break
        else:
            print("⚠️ Invalid choice. Try again.")

if __name__ == "__main__":
    main()
```
