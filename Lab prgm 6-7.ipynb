```python
from transformers import pipeline
# Zero Shot training
classifier = pipeline('zero-shot-classification', model='facebook/bart-large-mnli')

text = "Hugging Face is creating tools that democratize AI."
candidate_labels = ["technology", "education", "sports"]

result = classifier(text, candidate_labels)
print(result)
```

```python
from transformers import pipeline

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# One-shot prompt for Summarization
prompt = """
You are a summarization model. Below is an example of a text and its summary. Based on this example, summarize the next text.

Example:
Text: "The Mona Lisa is a famous painting by Leonardo da Vinci. It is considered one of the most famous pieces of art in the world. The painting is known for its unique style and mysterious expression."
Summary: "The Mona Lisa is a renowned painting by Leonardo da Vinci, known for its style and mysterious expression."

Now, summarize this text:
Text: "Hugging Face Inc. is a company based in New York City. Its headquarters are in DUMBO, very close to the Manhattan Bridge. Hugging Face is known for creating the 'transformers' library, a popular tool in the field of natural language processing. Their mission is to democratize AI and make it accessible to everyone."
"""

result = summarizer(prompt, max_length=50, min_length=25, do_sample=False)
print(result[0]['summary_text'])
```

### Lab Program 6:
### Title: Few-Shot Summarization using Prompt Engineering

```python
from transformers import pipeline

summarizer = pipeline("summarization", model="google/flan-t5-large")

# Few-shot prompt for Summarization
prompt = """
You are a summarization model. Below are examples of texts and their summaries. Based on these examples, summarize the next text.

Example 1:
Text: "The Mona Lisa is a famous painting by Leonardo da Vinci. It is considered one of the most famous pieces of art in the world. The painting is known for its unique style and mysterious expression."
Summary: "The Mona Lisa is a renowned painting by Leonardo da Vinci, known for its style and mysterious expression."

Example 2:
Text: "Apple Inc. is an American multinational technology company headquartered in Cupertino, California. It designs, manufactures, and sells consumer electronics, software, and online services. Apple's most well-known products include the iPhone, iPad, and Mac computers."
Summary: "Apple Inc. is a tech company known for products like the iPhone, iPad, and Mac computers."

Example 3:
Text: "The Amazon Rainforest, located in South America, is the largest tropical rainforest in the world. It plays a critical role in regulating the Earth's climate and is home to millions of species of plants and animals. However, deforestation is threatening its existence."
Summary: "The Amazon Rainforest is the largest tropical rainforest, crucial for regulating Earth's climate but threatened by deforestation."

Now, summarize this text:
Text: "Hugging Face Inc. is a company based in New York City. Its headquarters are in DUMBO, very close to the Manhattan Bridge. Hugging Face is known for creating the 'transformers' library, a popular tool in the field of natural language processing. Their mission is to democratize AI and make it accessible to everyone."
"""

result = summarizer(prompt, max_new_tokens=100)
print(result)
```

### Lab Program 7:
### Title: Structured vs Unstructured Prompts for Summarization Tasks

```python
from transformers import pipeline

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Structured Prompt for Summarization
prompt = """
You are a summarization model. Please summarize the following text:
"Hugging Face Inc. is a company based in New York City. Its headquarters are in DUMBO, very close to the Manhattan Bridge. Hugging Face is known for creating the 'transformers' library, a popular tool in the field of natural language processing. Their mission is to democratize AI and make it accessible to everyone."
"""

result = summarizer(prompt, max_length=50, min_length=25, do_sample=False)
print(result[0]['summary_text'])
```
