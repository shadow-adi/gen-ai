```python
# Few Shot Training and Prompt
import requests

API_URL = "https://router.huggingface.co/hf-inference/models/cardiffnlp/twitter-roberta-base-sentiment"
headers = {"Authorization": "Bearer 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"}

def query(payload):
  response = requests.post(API_URL, headers=headers, json=payload)
  return response.json()

# Enhanced prompt with multiple features (structured response, justification, few-shot learning, emphasis on key words)
input_text = """
You are a sentiment analysis model. Your task is to analyze the sentiment of text, and for each sentence, return:
1. The sentiment (positive, negative, or neutral).
2. The probability score for each class (positive, neutral, negative) in percentage.
3. A short explanation of why the sentiment was chosen, focusing on specific words or phrases.

Here are a few examples:
- Input: "I love this product, it's fantastic!"
  - Sentiment: Positive
  - Scores: Positive 90%, Neutral 5%, Negative 5%
  - Explanation: The word "love" expresses a strong positive emotion, and "fantastic" reinforces this positivity.

- Input: "This is the worst experience I've ever had."
  - Sentiment: Negative
  - Scores: Positive 1%, Neutral 5%, Negative 94%
  - Explanation: The phrase "worst experience" conveys strong dissatisfaction, making the sentiment negative.

Now, analyze the following text:
"I like it. I hate this movie."
"""

# Send the query to the model with the enhanced prompt
output = query({"inputs": input_text})

print(output)
```
